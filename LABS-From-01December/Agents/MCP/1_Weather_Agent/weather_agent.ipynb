# AGENT with MCP

MCP (Model Context Protocol) is an open-source standard for connecting AI applications to external systems.

Using MCP, LLM applications  can connect to data sources (e.g. local files, databases), tools (e.g. search engines, calculators) and workflows (e.g. specialized prompts)â€”enabling them to access key information and perform tasks.


 The Agents Python SDK understands multiple MCP transports. MCP gives LLLM - a single, unified interface to interact with a wide range of external services. This lets you reuse existing MCP servers or build your own to expose filesystem, HTTP, or connector backed tools to an agent.

MCP transports : 
- Hosted MCP server tools
- HTTP with SSE MCP servers
- stdio MCP servers

# Weather Agent using MCP
- Run a local wMCP server using FastMCP SSE , exposing weather tools over http://127.0.0.1:8000/sse
- Connect to the MCP server using Server-Sent Events (SSE) so that agent can access external weather tool.
- Run the agent with a query, prompting it to pick and execute the right tool.
- Print the final processed answer returned by the agent.

```python

#Connects to an MCP server using Server-Sent Events (SSE) so that agent can access external weather tool.
async with MCPServerSse(
    name="SSE Python Server",
    params={
        "url": "http://localhost:8000/sse",
        "headers": {"X-Workspace": workspace_id},
    },
    cache_tools_list=True, # Caches the tool list for faster performance and cleaner workflow.
) as server:
    agent = Agent(
        name="Assistant",
        mcp_servers=[server],
        model_settings=ModelSettings(tool_choice="required"), #  Cconfigured to always use a tool rather than guess answers.
    )
    result = await Runner.run(agent, "What's the weather in chennai?")
    print(result.final_output)
```
